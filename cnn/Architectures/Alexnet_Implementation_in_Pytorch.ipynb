# Implementation


```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
# 1Ô∏è Data Augmentation (Added RandomCrop & HorizontalFlip)
transform_augmented = transforms.Compose([
    transforms.RandomCrop(32, padding=4),       # Randomly crop within a 4-pixel boundary
    transforms.RandomHorizontalFlip(),          # Randomly flip images horizontally
    transforms.ToTensor(),                      # Convert to Tensor
    transforms.Normalize(mean=[0.4914, 0.4822, 0.4465], std=[0.2023, 0.1994, 0.2010])  # Standard CIFAR-10 normalization
])
```

```python
# Load CIFAR-10 with Augmentation
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_augmented)
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_augmented)

trainloader = torch.utils.data.DataLoader(trainset, batch_size=128, shuffle=True, num_workers=2)
testloader = torch.utils.data.DataLoader(testset, batch_size=100, shuffle=False, num_workers=2)
```

```python
#  AlexNet Model with Dropout
class AlexNet(nn.Module):
    def __init__(self, in_channels=3, classes=10):
        super(AlexNet, self).__init__()

        # Convolutional Layers
        self.conv1 = nn.Conv2d(in_channels, 96, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(96, 256, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(256, 384, kernel_size=3, stride=1, padding=1)
        self.conv4 = nn.Conv2d(384, 384, kernel_size=3, stride=1, padding=1)
        self.conv5 = nn.Conv2d(384, 256, kernel_size=3, stride=1, padding=1)

        # Fully Connected Layers (with Dropout)
        self.fc1 = nn.Linear(256 * 4 * 4, 4096)
        self.fc2 = nn.Linear(4096, 4096)
        self.fc3 = nn.Linear(4096, classes)

        # Activation, Pooling, Normalization, and Dropout
        self.relu = nn.ReLU()
        self.norm = nn.LocalResponseNorm(5)
        self.maxpool = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)
        self.dropout = nn.Dropout(0.3)  #  Added Dropout for Regularization

    def forward(self, x):
        # Convolutional + Activation + Normalization + MaxPooling
        x = self.maxpool(self.norm(self.relu(self.conv1(x))))
        x = self.maxpool(self.norm(self.relu(self.conv2(x))))
        x = self.relu(self.conv3(x))
        x = self.relu(self.conv4(x))
        x = self.maxpool(self.relu(self.conv5(x)))

        # Flatten before Fully Connected Layers
        x = torch.flatten(x, 1)

        # Fully Connected Layers with Dropout
        x = self.dropout(self.relu(self.fc1(x)))
        x = self.dropout(self.relu(self.fc2(x)))
        x = self.fc3(x)

        return F.log_softmax(x, dim=1)  # Softmax + Log for NLLLoss
```

```python
# 3 Initialize Model, Optimizer, Loss, and Scheduler
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = AlexNet().to(device)

#  Initialize Model, Optimizer, Loss
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9, weight_decay=0.0001)
criterion = nn.NLLLoss()
```

```python
# Dynamic LR Reduction Strategy
def adjust_learning_rate(optimizer, factor=0.1):
    for param_group in optimizer.param_groups:
        param_group['lr'] *= factor
    print(f"üîπ Learning rate reduced to {param_group['lr']:.5f}")
```

```python
# Modified Training Function
def train_model(model, trainloader, testloader, criterion, optimizer, num_epochs=30, early_stop_patience=5):
    train_losses, val_losses, accuracies = [], [], []
    best_val_loss = float('inf')
    plateau_count = 0  # Tracks stagnation to reduce LR

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for inputs, labels in trainloader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        avg_train_loss = running_loss / len(trainloader)
        train_losses.append(avg_train_loss)

        # Validation
        model.eval()
        val_loss, correct, total = 0.0, 0, 0
        with torch.no_grad():
            for inputs, labels in testloader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()

                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        avg_val_loss = val_loss / len(testloader)
        val_losses.append(avg_val_loss)

        accuracy = 100 * correct / total
        accuracies.append(accuracy)

        print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {avg_train_loss:.4f}, "
              f"Val Loss: {avg_val_loss:.4f}, Accuracy: {accuracy:.2f}%")

        # Learning Rate Adjustment Logic
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            plateau_count = 0
        else:
            plateau_count += 1
            if plateau_count >= early_stop_patience:
                adjust_learning_rate(optimizer)
                plateau_count = 0  # Reset plateau counter after LR reduction

    return train_losses, val_losses, accuracies
```

```python
train_losses, val_losses, accuracies = train_model(
    model, trainloader, testloader, criterion, optimizer, num_epochs=30, early_stop_patience=5
)
```

```python
#  Plot Results
plt.figure(figsize=(6, 4))
plt.plot(train_losses, label='Training Loss', color='blue')
plt.plot(val_losses, label='Validation Loss', color='red')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss (With Dropout & Augmentation)')
plt.legend()
plt.grid()
plt.show()
print("\n")
plt.figure(figsize=(6, 4))
plt.plot(accuracies, label='Validation Accuracy', color='green')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title('Validation Accuracy Over Time')
plt.legend()
plt.grid()
plt.show()
```
